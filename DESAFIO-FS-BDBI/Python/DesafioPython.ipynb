Desafio 02 - Agrupamento de Dados e Plotagem de Gráficos

```python
# Plote um gráfico de barras (bar) que exiba a quantidade de funcionários em cada setor (coluna Department);
df2 = df.groupby('Department')[['EmployeeCount']].sum()
df2
```

```python
# Plotagem gráfico de barras - funcionários x departamento
df2.plot(kind='bar', color='green',title='Quantidade de funcionários por Departamento')
```

O banco de dados aponta que 63 funcionários são do departamento de recursos humanos, 961 funcionários são do departamento de pesquisa e desenvolvimento e 446 funcionários são do departamento de vendas.

```python
# Plote um gráfico de linha (line) que exiba a quantidade de funcionários para cada nível de escolaridade (coluna Education);
df3 = df.groupby('Education')[['EmployeeCount']].sum()
df3
```

```python
# Plotagem gráfico de linhas - funcionários x escolaridade
df3.plot(kind='line', color='orange',title='Quantidade de funcionários por Escolaridade')
```

O banco de dados aponta que 170 funcionários possuem nivel de escolaridade tipo 1, 282 funcionários possuem nivel de escolaridade tipo 2, 572 funcionários possuem nivel de escolaridade tipo 3, 398 funcionários possuem nivel de escolaridade tipo 4 e 48 funcionários possuem nivel de escolaridade tipo 5.

```python
# Plote um gráfico de torta/pizza (pie) que exiba a quantidade de funcionários que tem ou não algum atrito na empresa (coluna Attrition).
df4 = df.groupby('Attrition')[['EmployeeCount']].sum()
df4
```

```python
# Plotagem gráfico de pizza - funcionários x atrito
labels = 'No', 'Yes'
sizes = [1233, 237]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels)
```

O banco de dados aponta que 1233 funcionários não possuem histórico de atrito na empresa, enquanto que 237 funcionários possuem histórico de atrito na empresa.
